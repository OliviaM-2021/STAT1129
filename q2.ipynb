```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# read in data
data = pd.read_csv('netflix_titles.csv')
```

```python
data
```

```python
# throw away missing values
data = data.dropna()
```

```python
data
```

```python
# get actors 
casts = data.cast.values
actor_freq = dict() 

for cast in casts:
    arr = cast.split(',')
    for name in arr:
        name = name.strip()
        actor_freq[name] = actor_freq.get(name, 0) + 1

top_actors = list(sorted(actor_freq.items(), key=lambda x: -x[1]))
print('There are', len(actor_freq), 'actors/actresses')
print('And TOP 5 are (name, appearence):', top_actors[:5])
```

```python
# Count of rating 
rating = data.rating.value_counts()
plt.figure()
plt.bar(rating.index, rating.values)
plt.xticks(ticks=np.arange(len(rating.index)),labels=rating.index, rotation='vertical')
plt.title('Count of Ratings')
plt.show()
```

```python
# get countries 
countries = data.country.values
country_freq = dict() 

for country in countries:
    arr = country.split(',')
    for c_name in arr:
        c_name = c_name.strip()
        country_freq[c_name] = country_freq.get(c_name, 0) + 1

top_countries = dict(sorted(country_freq.items(), key=lambda x: -x[1]))
print('There are', len(country_freq), 'countries')

n = 10
plt.figure()
plt.bar(list(top_countries.keys())[:n], list(top_countries.values())[:n])
plt.xticks(ticks=np.arange(n),labels=list(top_countries.keys())[:n], rotation='vertical')
plt.title('Top {} Countries in Term of Show Productions on Netflix'.format(n))
plt.show()
```

```python
# TV show with the most seasons and Movies with the longest duration 
print(data.type.value_counts())
print()

# name: duration
tv_shows = dict()
movies = dict()

# iterate through data frame by row (index location)
for i in range(len(data)):
    title = data.iloc[i].title
    duration = data.iloc[i].duration
    
    is_movie = True if "min" in duration else False
    duration = int(duration.split()[0])
    
    if is_movie:
        movies[title] = duration
    else:
        tv_shows[title] = duration

        
longest_movie = ''
longest_min = 0
for m in movies:
    if movies[m] > longest_min:
        longest_movie = m
        longest_min = movies[m]
print(longest_movie, 'has', longest_min, 'min and it is the longest movie on Netflix')

shortest_movie = longest_movie
shortest_min = longest_min
for m in movies:
    if movies[m] < shortest_min:
        shortest_movie = m
        shortest_min = movies[m]
print(shortest_movie, 'has', shortest_min, 'min and it is the shortest movie on Netflix')

plt.figure()
plt.hist(movies.values())
plt.xlabel('min')
plt.ylabel('count')
plt.title('Movie Length Distribution')
plt.show()

print('Most of the TV shows are 1 or 2 seasons in duration')
longest_tv = ''
longest_season = 0
for tv in tv_shows:
    if tv_shows[tv] > longest_season:
        longest_tv = tv
        longest_season = tv_shows[tv]
print(longest_tv, 'has', longest_season, 'seasons and it is the longest TV show on Netflix')

plt.figure()
plt.hist(tv_shows.values())
plt.xlabel('season(s)')
plt.ylabel('count')
plt.title('TV Show Duration Distribution')
plt.show()

```
